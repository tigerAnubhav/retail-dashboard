```python
#Import Library

# Import seaborn
import seaborn as sns

# Import pandas
import pandas as pd

# Import matplotlib
from matplotlib import pyplot as plt
```

```python
# Import dataset
dataset = pd.read_csv('../database/CollegeBasketballPlayers2022.csv')
```

```python
# Melihat banyaknya kolom dan baris pada dataset
dataset.shape
```

```python
# Melihat 5 data pertama
print(dataset.head())
```

```python
print(dataset.columns)


dataset = dataset.rename(columns = {'Unnamed: 64' : 'role'})


dataset = dataset.drop(columns =['num','type','pick','Unnamed: 65'])
```

```python
print(dataset.describe())
```

```python
print(dataset.info())
```

```python
dataset.dropna(inplace = True)
dataset = pd.DataFrame(dataset).reset_index()
dataset.drop(['index'], axis = 1, inplace = True)
dataset = dataset[['player_name','conf','yr','role','Min_per','pts']]
dataset.to_csv('../database/CollegeBasketballPlayers2022_clean.csv',index=False)
dataset.to_csv('../backend/basketball.csv',index=False)
```

```python
#Player Count per Conference
from collections import Counter
conference_count = dataset['conf'].to_list()
plt.figure(figsize=(10,10))
plt.pie(Counter(conference_count).values(), labels=Counter(conference_count).keys())
```
